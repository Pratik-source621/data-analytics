# Pandas â€“ Visualization

```python
import pandas as pd
import numpy as np
df=pd.DataFrame(np.random.randn(10,2),
                index=pd.date_range('1/1/2000',periods=10), 
                columns=list(['A','D']))
print(df)
df.plot()
```

```python
import pandas as pd
import numpy as np
df=pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])
print(df)
df.plot.bar()
```

```python
import pandas as pd
df=pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])
print(df)
df.plot.bar(stacked=True)
```

```python
import pandas as pd
import numpy as np
df=pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])
print(df)
df.plot.barh(stacked=True)
```

```python
import pandas as pd
import numpy as np
df=pd.DataFrame({'a':np.random.randn(1000)+8,'b':np.random.randn(1000),'c':np.random.randn(1000) - 3}, columns=['a', 'b', 'c'])
print(df)
df.plot.hist(bins=20)
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.rand(10, 5), columns=['A', 'B', 'C', 'D', 'E'])
print(df)
df.plot.box()
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.rand(10, 4), 
                  columns=['a', 'b', 'c', 'd'])
print(df)
df.plot.area()
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.rand(50, 4), columns=['a', 'b', 'c', 'd'])
print(df)
df.plot.scatter(x='a', y='d')
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(3 * np.random.rand(4), index=['a', 'b', 'c', 'd'], 
                  columns=['x'])
print(df)
df.plot.pie(subplots=True)
```
